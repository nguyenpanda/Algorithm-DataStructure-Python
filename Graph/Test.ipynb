```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a directed graph object
G = nx.DiGraph()

# Adding weighted edges to the graph
G.add_edge('A', 'B', weight=2)  # Edge from 'A' to 'B' with weight 2
G.add_edge('A', 'C', weight=1)  # Edge from 'A' to 'C' with weight 1
G.add_edge('B', 'C', weight=3)  # Edge from 'B' to 'C' with weight 3
G.add_edge('C', 'D', weight=4)  # Edge from 'C' to 'D' with weight 4
G.add_edge('D', 'A', weight=5)  # Edge from 'D' to 'A' with weight 5

# Draw the graph with edge labels
pos = nx.spring_layout(G, seed=42)  # Set seed for reproducibility
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=20, arrows=True)
labels = nx.get_edge_attributes(G, 'weight')  # Get edge weights as labels
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Weighted Directed Graph")
plt.axis('off')
plt.show()
```
