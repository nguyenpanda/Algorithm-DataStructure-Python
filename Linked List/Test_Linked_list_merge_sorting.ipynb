```python
from Linked_list_merge_sort import check, merge, merge_sort
from Linked_list import Node, SinglyLinkedList
from MyPackage.Tool import performance
import random

import numpy as np
from time import time
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def create_random_list(length: int=100, a=0, b=20):
    ls = SinglyLinkedList()
    
    for _ in range(length):
        random_number = random.randint(a, b)
        ls.prepend(random_number)
    
    return ls

def main(number: int=100):
    unsorted_list = []
    sorted_list = []
    list_length = []
    corrected_list = []
    time_list = []
    
    for i in range(number):
        linked_list_length = random.randint(0, 300)
        
        linked_list = create_random_list(linked_list_length, -300, 300)
        start = time()
        linked_list_sorted = merge_sort(linked_list)
        end = time()
        correct = check(linked_list_sorted)
        corrected_list.append(correct)
        time_list.append(end - start)
        
        if not correct:
            unsorted_list.append(linked_list)
            sorted_list.append(linked_list_sorted)
        
        if i % 100 == 0:
            print(f'Finished {i} case!')
        
        list_length.append(linked_list_length)
    return unsorted_list, sorted_list, corrected_list, time_list, list_length
```

```python
INPUT, OUTPUT, CHECK, TIME, LENGTH = main(200000)
```
